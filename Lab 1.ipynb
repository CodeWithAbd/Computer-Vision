# Transformations using OpenCV

## Scaling

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def resize_image(image_path, scale_factor):
    # Load the image
    img = cv.imread(image_path)
    
    # Resize the image using OpenCV
    res = cv.resize(img, None, fx=scale_factor, fy=scale_factor, interpolation=cv.INTER_CUBIC)

    # Show the resulting image
    plt.axis('off')
    plt.imshow(res)
    plt.show()



resize_image('img.jpeg', 2)
```

## Translation

```python
import numpy as np
import cv2 as cv

def load_image(image_path):
    """Loads an image from the given file path"""
    return cv.imread(image_path, 0)  # Load as grayscale

def affine_transform(image, x_shift, y_shift):
    """Applies an affine transformation to the given image"""
    rows, cols = image.shape
    M = np.float32([[1, 0, x_shift], [0, 1, y_shift]])
    return cv.warpAffine(image, M, (cols, rows))

def display_image(image):
    """Displays the given image"""
    cv.imshow('img', image)
    cv.waitKey(0)
    cv.destroyAllWindows()

# Load the image
img = load_image('img.jpeg')

# Apply an affine transformation
dst = affine_transform(img, 100, 50)

# Display the resulting image
display_image(dst)
```

## Rotation

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def load_image(image_path):
    """Loads an image from the given file path"""
    return cv.imread(image_path, 0)  # Load as grayscale

def rotate_image(image, angle):
    """Rotates the given image by the specified angle"""
    rows, cols = image.shape
    M = cv.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    return cv.warpAffine(image, M, (cols, rows))

def display_image(image):
    """Displays the given image"""
    plt.axis('off')
    plt.imshow(image)
    plt.show()

# Load the image
img = load_image('img.jpeg')

# Rotate the image
dst = rotate_image(img, 90)

# Display the resulting image
display_image(dst)
```

## Affine Transformation

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def load_image(image_path):
    """Loads an image from the given file path"""
    return cv.imread(image_path)

def apply_affine_transform(image, src_points, dst_points):
    """Applies an affine transformation to the given image using the specified source and destination points"""
    rows, cols, ch = image.shape
    M = cv.getAffineTransform(src_points, dst_points)
    return cv.warpAffine(image, M, (cols, rows))

def display_images(input_image, output_image):
    """Displays the input and output images side by side"""
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(input_image)
    ax1.set_title('Input')
    ax2.imshow(output_image)
    ax2.set_title('Output')
    plt.show()

# Load the image
img = load_image('img.jpeg')

# Define the source and destination points
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])

# Apply an affine transformation to the image
dst = apply_affine_transform(img, pts1, pts2)

# Display the input and output images side by side
display_images(img, dst)
```

## Perspective Transformation

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def load_image(image_path):
    """Loads an image from the given file path"""
    return cv.imread(image_path)

def apply_perspective_transform(image, src_points, dst_points, output_size):
    """Applies a perspective transformation to the given image using the specified source and destination points"""
    M = cv.getPerspectiveTransform(src_points, dst_points)
    return cv.warpPerspective(image, M, output_size)

def display_images(input_image, output_image):
    """Displays the input and output images side by side"""
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(input_image)
    ax1.set_title('Input')
    ax2.imshow(output_image)
    ax2.set_title('Output')
    plt.show()

# Load the image
img = load_image('img.jpeg')

# Define the source and destination points
pts1 = np.float32([[56, 65], [368, 52], [28, 387], [389, 390]])
pts2 = np.float32([[0, 0], [300, 0], [0, 300], [300, 300]])

# Apply a perspective transformation to the image
dst = apply_perspective_transform(img, pts1, pts2, (300, 300))

# Display the input and output images side by side
display_images(img, dst)
```

## Geometric Transformation 

```python
import cv2
import numpy as np

img = cv2.imread('img.webp')
num_rows, num_cols = img.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 30, 1)
img_rotation = cv2.warpAffine(img, rotation_matrix, (num_cols, num_rows))

plt.axis('off')
plt.imshow(img_rotation)
plt.show()
```
